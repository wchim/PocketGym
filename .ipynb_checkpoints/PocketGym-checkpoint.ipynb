```python
import pandas as pd

stats = pd.read_csv("stats.csv")
stats.head()
```

```python
stats.dtypes
```

```python
#Bench Press Leaderboard
bench = stats.copy()
bench = bench[["member_id",
               "first_name",
               "last_name",
               "body_weight",
               "bench_press"]]
bench["bp:bw"] = round((bench.iloc[:,4]/bench.iloc[:,3]),2)
bench = bench.sort_values("bp:bw", ascending=False)
bench.reset_index(drop=True, inplace=True)
bench.head()
```

```python
#Deadlift Leaderboard
lift = stats.copy()
lift = lift[["member_id",
             "first_name",
             "last_name",
             "body_weight",
             "deadlift"]]
lift["dl:bw"] = round((lift.iloc[:,4]/lift.iloc[:,3]),2)
lift = lift.sort_values("dl:bw", ascending=False)
lift.reset_index(drop=True, inplace=True)
lift.head()
```

```python
#Back Squat Leaderboard
squat = stats.copy()
squat = squat[["member_id",
               "first_name",
               "last_name",
               "body_weight",
               "back_squat"]]
squat["bs:bw"] = round((squat.iloc[:,4]/squat.iloc[:,3]),2)
squat = squat.sort_values("bs:bw", ascending=False)
squat.reset_index(drop=True, inplace=True)
squat.head()
```

```python
#Military Press Leaderboard
military = stats.copy()
military = military[["member_id",
                  "first_name",
                  "last_name",
                  "body_weight",
                  "military_press"]]
military["mp:bw"] = round((military.iloc[:,4]/military.iloc[:,3]),2)
military = military.sort_values("mp:bw", ascending=False)
military.reset_index(drop=True, inplace=True)
military.head()
```

```python
def calculate(member_id, exercise, perc):
    orm = stats.iloc[member_id][exercise]
    adjusted_orm = 0.9*orm
    
    temp_total_weight = perc*adjusted_orm
    
    side_weights = (temp_total_weight - 45)/2
    
    if side_weights % 5.0 > side_weights % 10.0:
        side_weights = side_weights - side_weights % 10.0
    else:
        side_weights = side_weights - side_weights % 5.0
    side_weights = round(side_weights, 0)
    
    act_total_weight = side_weights*2 + 45
    
    this_set = f"Total Weight: {act_total_weight} " \
               f"Weight per Side: {side_weights}"
    print(this_set)
    return None
```

```python
exercises = ["bench_press", "deadlift", "back_squat", "military_press"]

def routine(member_id, week, exercise):
    print(stats.loc[stats['member_id'] == 3, 'first_name'])
    if week%4 == 0:
        print("Reps: 5, 5, 5")
        for exercise in exercises:
            print(exercise)
            calculate(member_id, exercise, .4)
            calculate(member_id, exercise, .5)
            calculate(member_id, exercise, .6)
    elif week%3 == 0:
        print("Reps: 5, 3, 1+")
        for exercise in exercises:
            print(exercise)
            calculate(member_id, exercise, .75)
            calculate(member_id, exercise, .85)
            calculate(member_id, exercise, .96)
    elif week%2 == 0:
        print("Reps: 3, 3, 3+")
        for exercise in exercises:
            print(exercise)
            calculate(member_id, exercise, .7)
            calculate(member_id, exercise, .8)
            calculate(member_id, exercise, .9)
    else:
        print("Reps: 5, 5, 5+")
        for exercise in exercises:
            print(exercise)
            calculate(member_id, exercise, .65)
            calculate(member_id, exercise, .75)
            calculate(member_id, exercise, .86)
```

```python
for num in range(4):
    routine(3, num, exercises)
```

```python
stats.to_csv('stats.csv', index=False)
bench.to_csv('bench_ranking.csv', index=False)
lift.to_csv('lift_ranking.csv', index=False)
squat.to_csv('squat_ranking.csv', index=False)
military.to_csv('military_ranking.csv', index=False)
```
