```python
import pandas as pd

stats = pd.read_csv("stats.csv")
stats.head()
# stats = pd.DataFrame(columns=("first_name",
#                               "last_name",
#                               "height",
#                               "body_weight",
#                               "bench_press",
#                               "deadlift",
#                               "back_squat",
#                               "military_press"))
```

```python
person = []

for feature in stats.columns:
    info = input(f"Enter your {feature}: ")
    person.append(info)

person = pd.Series(person, index=stats.columns)
person
stats = stats.append(person, ignore_index=True)
```

```python
#Bench Press Leaderboard
bench = stats[["member_id",
               "first_name",
               "last_name",
               "body_weight",
               "bench_press"]]
bench["bp:bw"] = round(stats["bench_press"]/stats["body_weight"],2)
bench = bench.sort_values("bp:bw", ascending=False)
bench.head()
```

```python
#Deadlift Leaderboard
lift = stats[["member_id",
               "first_name",
               "last_name",
               "body_weight",
               "deadlift"]]
lift["dl:bw"] = round(stats["deadlift"]/stats["body_weight"],2)
lift = lift.sort_values("dl:bw", ascending=False)
lift.head()
```

```python
#Back Squat Leaderboard
squat = stats[["member_id",
               "first_name",
               "last_name",
               "body_weight",
               "back_squat"]]
squat["bs:bw"] = round(stats["back_squat"]/stats["body_weight"],2)
squat = squat.sort_values("bs:bw", ascending=False)
squat.head()
```

```python
#Military Press Leaderboard
military = stats[["member_id",
               "first_name",
               "last_name",
               "body_weight",
               "military_press"]]
military["mp:bw"] = round(stats["military_press"]/stats["body_weight"],2)
military = military.sort_values("mp:bw", ascending=False)
military.head()
```

```python
def calculate(member_id, exercise, perc):
    orm = stats.iloc[member_id][exercise]
    adjusted_orm = 0.9*orm
    
    temp_total_weight = perc*adjusted_orm
    
    side_weights = (temp_total_weight - 45)/2
    
    if side_weights % 5.0 > side_weights % 10.0:
        side_weights = side_weights - side_weights % 10.0
    else:
        side_weights = side_weights - side_weights % 5.0
    side_weights = round(side_weights, 0)
    
    act_total_weight = side_weights*2 + 45
    
    this_set = f"Total Weight: {act_total_weight} " \
               f"Weight per Side: {side_weights}"
    print(this_set)
    return None
```

```python
exercises = ["bench_press", "deadlift", "back_squat", "military_press"]

print(stats.iloc[0]["first_name"])
for perc in range(40,100,5):
    print(f"{perc}%")
    for exercise in exercises:
        print(exercise)
        calculate(0, exercise, perc/100)
```

```python
stats.to_csv('stats.csv')
bench.to_csv('bench_ranking.csv')
lift.to_csv('lift_ranking.csv')
squat.to_csv('squat_ranking.csv')
military.to_csv('military_ranking.csv')
```
